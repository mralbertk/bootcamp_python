# Advanced Iteration

In data science, we often need to move beyond simple loops to perform high-level operations on entire datasets. These tools are highly efficient and are considered "Pythonic" standard practice.

## Boolean Checkers: `any()` and `all()`

These functions allow you to evaluate logical conditions across an entire collection without writing a manual loop with a counter.

* **`any(iterable)`**: Returns `True` if **at least one** element in the iterable is true.
* **`all(iterable)`**: Returns `True` only if **every** element in the iterable is true.

In practice, we usually pass a "generator expression" (a comprehension without the brackets) to these functions to check a specific property.

```python
results = [0.98, 0.95, 0.88, 0.92, 1.05] # One value is an outlier (> 1.0)

# Are there any outliers?
has_outliers = any(x > 1.0 for x in results) 
print(f"Has outliers: {has_outliers}") # True

# Are all results valid (non-negative)?
all_valid = all(x >= 0 for x in results)
print(f"All valid: {all_valid}") # True
```

## Pairing Iterables: `zip()`

The `zip()` function takes two or more sequences and "zips" them together into a single iterable of tuples. It stops as soon as the shortest input sequence is exhausted.

Mathematically, if you have vectors $\mathbf{a}$ and $\mathbf{b}$, `zip` produces the sequence of pairs $(a_0, b_0), (a_1, b_1), \dots, (a_n, b_n)$.

```python
headers = ["ID", "Score", "Status"]
values = [101, 95.5, "Pass"]

# Create a list of pairs
zipped = list(zip(headers, values))
print(zipped) # [('ID', 101), ('Score', 95.5), ('Status', 'Pass')]

# Common use case: Creating a dictionary from two lists
data_dict = dict(zip(headers, values))
```

## Transformations: `map()` and `filter()`

These two functions represent the core of "functional" data processing. 

### `map(function, iterable)`
Applies a specific function to every item in an iterable. 

```python
def normalize(x):
    return x / 100

scores = [85, 92, 78, 90]
normalized_scores = list(map(normalize, scores))
print(normalized_scores) # [0.85, 0.92, 0.78, 0.9]
```

### `filter(function, iterable)`
Constructs an iterator from those elements of the iterable for which the function returns `True`.

```python
def is_passing(score):
    return score >= 60

grades = [85, 42, 77, 55, 91]
passing_grades = list(filter(is_passing, grades))
print(passing_grades) # [85, 77, 91]
```

> **Note on Iterators:** `map`, `filter`, and `zip` all return **iterators**. This means they don't calculate the results until you actually "ask" for them (e.g., by wrapping them in `list()` or looping over them). This makes them very memory-efficient for large datasets.

# Generators

A generator is a special type of function that returns an **iterator** object. Unlike regular functions that calculate all results and return them in a list, a generator "yields" values one by one.

## The `yield` Keyword
The defining feature of a generator function is the `yield` statement. 
- When a function calls `return`, it is finished and its local state is destroyed.
- When a function calls `yield`, it **pauses** execution, saves its state, and gives a value back to the caller. When the caller asks for the next value, the function resumes exactly where it left off.

```python
def count_to_three():
    yield 1
    yield 2
    yield 3

# Creating the generator object
counter = count_to_three()

print(next(counter)) # 1
print(next(counter)) # 2
print(next(counter)) # 3
# print(next(counter)) # This would raise a StopIteration error
```

## Lazy Evaluation and Memory Efficiency
The most powerful aspect of generators is **Lazy Evaluation**. Values are not computed until the moment they are needed.

Imagine you need to process the squares of one million numbers.
- **A List** would calculate all 1,000,000 squares immediately and store them in RAM.
- **A Generator** stores only the instruction on how to calculate the next square.

```python
def square_generator(n):
    for i in range(n):
        yield i ** 2

# This does NOT calculate any squares yet. 
# It only creates a generator object.
big_squares = square_generator(1000000)

# We can iterate over it just like a list
for val in big_squares:
    if val > 100:
        break
    print(val)
```

## Infinite Sequences
Because generators are lazy, they can represent **infinite mathematical sequences** that would be impossible to store in a list.

```python
def infinite_powers_of_two():
    exponent = 0
    while True:
        yield 2 ** exponent
        exponent += 1

powers = infinite_powers_of_two()
print(next(powers)) # 1
print(next(powers)) # 2
print(next(powers)) # 4
# ... this can go on
```
