# Conditional Execution

By default, Python executes code line-by-line from top to bottom. Conditional execution allows us to alter this "flow" based on whether a statement is `True` or `False`.

## Basic Conditionals: `if`, `elif`, and `else`

The `if` statement evaluates a condition (a boolean expression). If the condition is `True`, the indented block of code below it runs.

* **`if`**: The initial check.
* **`elif`** (short for else-if): Allows you to check multiple alternative conditions if the previous ones were `False`.
* **`else`**: The "catch-all" block that runs if none of the above conditions were met.

```python
def check_temperature(temp):
    if temp > 30:
        print("It is hot.")
    elif temp >= 15:
        print("It is mild.")
    else:
        print("It is cold.")

check_temperature(25) # Output: It is mild.
```

**Important Syntax Note:** 1.  The condition must end with a **colon** (`:`).
2.  The code to be executed must be **indented**.

## Structured Pattern Matching: `match`

The `match` statement is a cleaner way to compare a single variable against many specific values. It functions similarly to a "switch" statement in other languages.

```python
def handle_status(status_code):
    match status_code:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown Status"

print(handle_status(404)) # Output: Not Found
```
*Note: The underscore `_` acts as a wildcard (like `else`), matching anything that wasn't caught by previous cases.*

## Avoiding Crashes: `try` and `except`

In data, your code often encounters "dirty" data, for example a string where a number was expected. If Python tries to perform an operation it can't handle (e.g., dividing by zero), the program **crashes** with an "Exception."

We use `try` and `except` blocks to handle these errors gracefully without stopping the entire program.

```python
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: You cannot divide by zero!"
    except TypeError:
        return "Error: Please provide numeric inputs."

print(safe_divide(10, 0))    # Output: Error: You cannot divide by zero!
print(safe_divide(10, "2"))  # Output: Error: Please provide numeric inputs.
```
