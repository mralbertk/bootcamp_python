# Unordered Collections
In a sequence, the order is part of the data's meaning. In an unordered collection, the focus is on the **relationship** between data points or the **uniqueness** of the data itself.

## Sets (`set`)
A **Set** is an unordered collection of **unique** elements. It is the Python implementation of the mathematical set ($\mathbb{S}$).

### Key Properties:
* **No Duplicates:** Adding an existing element to a set has no effect.
* **Unordered:** You cannot access elements by index (e.g., `my_set[0]` will raise an error).
* **Highly Optimized:** Checking if an item exists in a set (`item in my_set`) is significantly faster than checking a list, especially for large datasets ($O(1)$ vs $O(n)$ complexity).

```python
# Creating sets
primes = {2, 3, 5, 7}
observations = set([1, 2, 2, 3, 3, 3]) # Result: {1, 2, 3}

# Membership testing
print(2 in primes)  # True
print(4 in primes)  # False
```

### Mathematical Set Operations
Python sets support standard operations from set theory, which are incredibly useful for comparing different groups of data.

| Operation | Operator | Method | Mathematical Symbol |
| :--- | :--- | :--- | :--- |
| **Union** | `\|` | `.union()` | $A \cup B$ |
| **Intersection** | `&` | `.intersection()` | $A \cap B$ |
| **Difference** | `-` | `.difference()` | $A \setminus B$ |

## Dictionaries (`dict`)
A **Dictionary** is a collection of **key-value pairs**. It represents a mapping $f: \text{Key} \to \text{Value}$.

### Key Rules:
1.  **Keys must be unique:** A dictionary cannot have two identical keys.
2.  **Keys must be immutable:** You can use strings, numbers, or tuples as keys, but **not lists**.
3.  **Values can be anything:** Values can be duplicates and of any data type (including other dictionaries or lists).

```python
# Defining a dictionary
student = {
    "name": "Alice",
    "age": 25,
    "courses": ["Math", "Python"]
}

# Accessing values (using square brackets)
print(student["name"])  # Alice

# Adding or updating
student["age"] = 26
student["graduated"] = False
```

### Iterating over Dictionaries
Since dictionaries have two components, we have three ways to iterate over them:

```python
# 1. Iterate over keys (default)
for key in student:
    print(key)

# 2. Iterate over values
for val in student.values():
    print(val)

# 3. Iterate over both (most common)
for key, val in student.items():
    print(f"{key}: {val}")
```

## Set and Dictionary Comprehensions

Just as with lists, Python provides a shorthand syntax to create sets and dictionaries from other iterables. For data science, this is extremely useful for transforming keys, filtering unique values, or "reversing" a mapping.

### Set Comprehensions
A set comprehension looks exactly like a list comprehension but uses **curly braces `{ }`**. Because it creates a set, any duplicate results are automatically discarded.

**Syntax:** `{expression for item in iterable if condition}`

```python
# Example: Extract unique squared values from a list containing duplicates
observations = [-2, -1, 1, 2, 3, -3]

# The set will automatically handle the fact that (-2)**2 and (2)**2 are both 4
unique_squares = {x**2 for x in observations}

print(unique_squares)  # Output: {1, 4, 9}
```

### Dictionary Comprehensions
Dictionary comprehensions also use curly braces, but they require a **key-value pair** separated by a colon (`key: value`).

**Syntax:** `{key_expression: value_expression for item in iterable if condition}`

```python
# Example 1: Create a mapping of numbers to their squares
square_map = {x: x**2 for x in range(1, 6)}
print(square_map)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Example 2: Swapping Keys and Values
# Useful in data science when you need a "reverse lookup"
original_data = {"ID_101": "Active", "ID_102": "Inactive", "ID_103": "Active"}

# Create a new dict where we only keep the 'Active' IDs and use the ID as the value
```
