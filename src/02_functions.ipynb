# Functions
A function is a named, reusable block of code designed to perform a specific task. 

For example, we may need to perform a data cleaning step on every row in a large data set. Instead of spelling out this step over and over again, we write it once and then call it as we process each line.

## Defining and Calling Functions
In Python, functions are defined using the `def` keyword. Every function definition consists of two parts:

- **The signature**: The name of the function and its input arguments
- **The body**: The code block that is executed when the function is called

````python
# Defining a function
def greet():
    print("hello world")
````

````python
# Calling a function
greet()
````

### Code Blocks
A code block is a unit of code that is executed together. In many programming languages, code blocks are enclosed by curly braces `{}`. In Python, a code block is defined by **indentation**, usually four spaces. One level of indentation indicates one coherend block of logic.

### Arguments and Return Values
To make a function useful, we often send it data (arguments) and get a result back (return value). Some important terminology to keep in mind:

- **Parameters:** Placeholders for data defined in the signature of the function.
- **Arguments:** The actual data passed to the function when calling it.
- **Return:** Use the `return` keyword to send data back to the main program.

Note that the `return` keyword always terminates the function's execution.

````python
def add_one(num):
    incremented_num = num + 1
    return incremented_num
````

````python
my_num = input("Enter a number: ")
my_num = int(my_num)
print(add_one(my_num))
````

Python offers flexibility in how you can pass arguments to functions:

- **Positional arguments** are the default. They are matched by the order they appear.
- **Keyword arguments** are explicitly named parameters.
- **Optional arguments** have a default value that is used when they are absent in a function call.

````python
def multiply(x, y):
    return x * y
````

````python
def area(height, width):
    return height * width
````

````python
def exponentiate(base, exponent=2):
    return base ** exponent
````

````python
# Calling the multiply() functions with positional arguments
multiply(4,2)
````

````python
# Calling the multiply() function with keyword arguments
multiply(y=2, x=4
````

````python
# Calling area() with keyword arguments
area(width=2, height=4)
````

````python
# Calling area() with positional arguments
area(2, 4)
````

````python
# Calling exponentiate() with one argument
exponentiate(3)
````

````python
# Calling exponentiate() with two arguments
exponentiate(3, 3)
````

## Scope: Local vs Global
Think of functions as programs inside a program. When a function is called, the execution of the main program is halted until the function finishes execution. 

One important consequence is that variables defined inside a function only exist within that function. They are destroyed when the function finishes execution. The `return` keyword is necessary to make this data available to the main function.

```python
def increment(num, increment_by = 1):
    incremented = num + increment_by  # Local scope: Only available in the function
    return incremented

my_incremented_number = increment(4, 2)
print(my_incremented_number)  # 6
print(incremented)            # Error!
```

Variables assigned outside your function - as part of your main program - are available everywhere. You can read data in a function that is not passed as an argument. You can even manipulate it, if you use the `global` keyword inside your function. However, this is a frequent source of bugs and generally not recommended, unless you *really* know what you're doing!
