# Iteration
Iteration allows us to process elements of a collection one by one. In Python, any object capable of returning its members one at a time is called an **iterable**. All the sequence types we have studied (lists, tuples, and strings) are iterables.

## The `for` Loop
The `for` loop is the most common way to iterate over a sequence. It assigns the current element to a variable and executes the code block for that element.

```python
observations = [10.2, 11.5, 9.8, 12.1]

# "val" is a placeholder for the current element
for val in observations:
    print(f"Processing observation: {val}")
```

### Iterating with `range()`
If you need to iterate a specific number of times or access elements by their index, use the `range()` function.

```python
# Iterating over indices
for i in range(len(observations)):
    print(f"Index {i} holds value {observations[i]}")
```

## The `enumerate()` Function
Sometimes you need both the **index** and the **value** of the items in a sequence. Instead of using `range(len(...))`, the "Pythonic" way is to use `enumerate()`. It returns a sequence of tuples containing `(index, value)`.

```python
# Unpacking index and value simultaneously
for idx, val in enumerate(observations):
    if val > 11.0:
        print(f"High value found at index {idx}: {val}")
```

## List Comprehensions
List comprehensions provide a concise way to create new lists based on existing iterables. For mathematicians, this syntax is very similar to **Set-builder notation**.

**Mathematical Notation:**
$$S = \{x^2 \mid x \in \{1 \dots 5\}\}$$

**Python List Comprehension:**
```python
# [expression for item in iterable if condition]
numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]

print(squares)  # [1, 4, 9, 16, 25]

# Adding a condition (filter)
even_squares = [x**2 for x in numbers if x % 2 == 0]
print(even_squares)  # [4, 16]
```

## The `while` Loop
A `while` loop repeats a block of code as long as a specific **condition** remains `True`. Use this when you do not know in advance how many iterations are required (e.g., waiting for a mathematical algorithm to converge).

```python
# Finding the first power of 2 greater than 1000
value = 1
while value < 1000:
    value = value * 2

print(f"The first power of 2 above 1000 is: {value}")
```

> **Warning:** Ensure your `while` loop has a condition that eventually becomes `False`, otherwise you will create an **infinite loop**, which will cause your program to hang.
