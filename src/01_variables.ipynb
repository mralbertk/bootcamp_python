# Variables
In programming, a variable is a **name** that refers to a **value** stored in the computer's memory. You can think of it as a label attached to an object.

In Python, we use the assignment operator `=` to bind a name to a value.

```python
# Assignment: 'x' now points to the integer object 10
x = 10

# Incrementing: This makes sense in code, but would be an impossible equation (x = x + 1)
x = x + 1
```

## Naming rules
The following rules apply to variable names:

- Allowed are upper- and lowercase letters, numbers, and underscores
- A name must begin with a letter or underscore
- Python keywords are forbidden (i.e. `if`, `else`, `def` ...)

Variable names are case sensitive. For example, these two names are different: `alpha` and `Alpha`

In addition to these hard rules there are also conventions you *should* respect:

- Use **snake_case**, lowercase words separated by underscore
- Use all uppercase for constants, variables that are not supposed to change
- Use an underscore prefix `_` to indicate a variable is for internal use only

Lastly, be careful not to use the name of built-in functions for variables. This will overwrite the original binding and make the functionality inaccessible.

For example, don't do this:

```python
sum = 4 + 2  # Now the built-in sum() function is inaccessible in your program.
```

## Typing
Objects in Python have a value and a type. The type informs Python what operations are permitted on this object.

```python
answer = 42             # A whole number (integer type)
answer = "Forty-two"    # Text  (string type)
```

Python is a **dynamically typed** language. This means:

- You do not need to state the type of an object when binding it to a variable
- You may re-assign variables to objects of another type during your program's execution

You can check the type of a variable with the `type()` or the `isinstance()` functions:

```python
type(answer)                # <class 'str'>
isinstance(answer, int)     # False
```

## Common Data Types
These are core data types you encounter frequently in Python:

| Type    | Description                 | Example                    |
| ------- | --------------------------- | -------------------------- |
| `int`   | Integer: Whole numbers      | `my_int = 42`              |
| `float` | Floating-point number       | `pi = 3.1415`              |
| `str`   | String (text)               | `greeting = "hello world"` |
| `bool`  | Boolean (`True` or `False`) | `is_valid = True`          |

## Referencing variables in text
`f-strings` are the modern Python way to use variables in text.

```python
name = "Alice"
print(f"Hello {name}!")
```

```python
pi = 3.1415
print(f"The value of pi is approximately {pi}.")
```

## Type casting
Python infers the type of a variable from the assignment operation.

```python
whole_num = 42      # Integer type
decimal_n = 42.0    # Float type
written_n = "42"    # String type
```

You can explicity override the inferred type using typecasting.

```python
whole_num = float(42)
type(whole_num)         # <class 'float'>
```

```python
written_n = int(written_n)
isinstance(written_n, int)  # True
```

Note that typecasting creates a new object in memory and re-binds your variable.
