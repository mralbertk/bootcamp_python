# Numeric Types
Python has four primary built-in numeric types:

| Type      | Description            | Example           |
| --------- | ---------------------- | ----------------- |
| `int`     | Integer: Whole numbers | `answer = 42`     |
| `float`   | Floating-point number  | `pi = 3.14159`    |
| `complex` | Complex number         | `z = 3 + 4j`      |
| `bool`    | Boolean: True or False | `is_valid = True` |

Remember: All objects in Python have a value and a type. The type informs Python what operations are permitted.

## Arithmetic Operations
Python supports standard arithmetic symbols. Pay attention to the resulting types.

| Symbol | Operation        | Example (`a = 3`, `b = 2`) | Result |
| ------ | ---------------- | ---------------------------| ------ |
| `+`    | Addition         | `a + b`                    | `5`    |
| `-`    | Subtraction      | `a - b`                    | `1`    |
| `*`    | Multiplication   | `a * b`                    | `6`    |
| `/`    | Division         | `a / b`                    | `1.5`  |
| `//`   | Integer Division | `a // b`                   | `1`    |
| `%`    | Modulo           | `a % b`                    | `1`    |
| `**`   | Exponentiation   | `a ** b`                   | `9`    |

Important:

- The outcome of a division is always a `float` type
- Integer division in Python uses f-division (as opposed to t-divison)

## Comparison Operations
These operators allow you to evaluate the relationship between two values. These operations always return a Boolean result (`True` or `False`).

| Symbol | Operation | Example (`a=3`, `b=2`) | Result |
| ---- | --------------------- | -------- | ---------- |
| `<`  | Strictly less than    | `a < b`  | `False`    |
| `>`  | Strictly greater than | `a > b`  | `True`     |
| `==` | Equal                 | `a == b` | `False`    |
| `!=` | Not equal             | `a != b` | `True`     | 
| `<=` | Less than or equal    | `a <= b` | `False`    |
| `>=` | Greater than or equal | `a >= b` | `True`     |

## Logical Operators
Use logical operators to combine multiple comparison statements.

- `and`: Returns `True` if all conditions are true.
- `or`: Returns `True` if at least one condition is true.
- `not`: Negates the condition
