# Recommendations to Go Further
As we said at the beginning: Becoming "good" at programming takes time and practice. 

Here are my personal recommendations for your continued learning, should you be so inclined. 

These are **not** part of the official DSTI curriculum, and also not part of your upcoming exam.

## General Practice

Getting used to syntax is just the first step. The real challenge is to learn how to *"think like a programmer"*. To get there, you need to solve problems. These sites make it a game:

- **[Advent of Code](https://adventofcode.com/):** A yearly series of logic puzzles; great for testing your control flow and data structure skills.
- **[Project Euler](https://projecteuler.net/):** Ideal for those with math backgrounds; these challenges require mathematical insights combined with efficient code.

## Future Data Scientists: The Mathematical Foundation

Data science in Python is built on the ability to perform high-speed numerical operations and manage complex data structures. These two libraries are essential to the task:

- **[Numpy](https://numpy.org/learn/):** The central library for scientific computation. It is the bedrock of almost every other data science tool. Understanding NumPy is essential for handling the high-performance math required for machine learning.
- **[Pandas](https://pandas.pydata.org/docs/getting_started/index.html):** A fantastic tool for data manipulation. It introduces the "DataFrame," which allows you to clean, merge, and organize datasets with a level of control that spreadsheets simply cannot match. 

## Future Data Analysts: Insight and Communication

Analytics is the art of extracting a narrative from raw data and ensuring that narrative is conveyed clearly to others.

- **[Pandas](https://pandas.pydata.org/docs/getting_started/index.html):** For data analysts, Pandas goes far beyond cleaning. Learning to use it for advanced aggregation, grouping, and pivoting is what allows you to find the "why" hidden in the numbers.
- **[Seaborn](https://seaborn.pydata.org/tutorial.html):** The best analysis is worthless if it isn't communicated well. Seaborn is a powerful library for data visualization, helping you create sophisticated graphics that make your findings persuasive and clear.

## Future Data Engineers: Systems and Delivery

Engineering focuses on building the "plumbing"—the systems that make data accessible and useful to users and applications.

- **[FastAPI](https://fastapi.tiangolo.com/learn/):** A high-performance tool for building APIs. Learning how to spin up an API is a critical skill for making your data and models accessible to the wider world through web services.
- **[Streamlit](https://docs.streamlit.io/):** This allows you to build interactive data applications quickly. It is incredibly valuable for prototyping "data products"—turning a simple script into a functional dashboard that others can interact with.
