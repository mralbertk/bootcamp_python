# Sequence Types

A sequence is an ordered collection of items. Each item is assigned a position, called an index. Python categorizes sequences in two main groups, based on whether they can be changed after creation:

- **Mutable Sequences**: Elements can be modified, added, or removed.
- **Immutable Sequences**: Elements are fixed once the sequence is created.

The primary mutable sequence we need to know is the **List**. Common immutable sequences are **Tuples** and **Strings**. 

## The `list` Type
The primary mutable sequence type in Python. It can hold items of different data types, even other lists. 

Lists are created using square brackets `[]` in an assignment operation, or via the `list()` constructor:

```python
my_list = []                  # Creates an empty list
my_other_list = list()        # Also creates an empty list
my_third_list = [1, "b", 3]   # Creates a list with the elements 1, "b", and 3
```

### Accessing Elements
Elements of the list are accessible via their index. The notation is `my_list[idx]`, where `idx` is the position of the element.

```python
print(my_third_list[1])
> "b"
```

Sequences in Python are **zero-indexed**, meaning the first element is at position 0. The last element is at position `length of list - 1`. To access the last element of a list, Python offers _negative indexing_ for convenience:

```python
print(my_third_list[-1])
> 3
```

*Side note: Use the built-in `len()` function to quickly get the number of elements in a sequence, i.e. `len(my_third_list)` returns `3`.* 

Attempting to access an element of a list that does not exist raises an `IndexError`:

```python
print(my_third_list[3])
> IndexError: list index out of range
```

We can access multiple elements of a list. This is called `slicing`. The notation to get a slice is the following:

```python
list[start:stop:step]
```

- `start` marks the index of the first element of the slice
- `stop` marks the index of the last but one element of the slice
- `step` marks the increment of the index for each element in the slice

```python
slice_me = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(slice_me[0:6:2])  # 1, 3, 5
```

Note that the `stop` value is exclusive. You can read the slicing notation as "from position x up to but not including position y by step of z".

If you omit a value, the default is used:

- `start` defaults to position `0`, the beginning of the list
- `stop` defaults to the end of the list, including the final element
- `step` defaults to 1, so collecting each element

```python
slice_me = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(slice_me[::])   # Full list
print(slice_me[2::])  # 3, 4, 5, 6, 7, 8, 9
print(slice_me[:5:2]) # 1, 3, 5
```

### Modifying Elements
The `list` is a mutable sequence. We can modify elements as we see fit:

```python
# Change an element 
my_list = ["a", "b", "c"]
my_list[2] = "d"
print(my_list)
> ["a", "b", "d"]
```

```python
# Add an element to the list
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)
> [1, 2, 3, 4]
```

## The `tuple` Type
Tuples are the primary immutable datatype in Python. Like the `list` type, a `tuple` can hold elements of different data types, including `list` and `tuple`. 

Tuples are created using parentheses `()`.

```python
my_tuple = (6, 7)
```

Elements in a tuple are accessible by their index.

```python
print(my_tuple[1])
> 7
```

Tuples do not allow modifications. There is no `.append()` method. Attempting to re-assign a value in a tuple raises an error.



# References and Object Identity
In Python, a variable is not a container that "holds" data; it is a **name** (or label) bound to an **object** in memory.

When you assign one variable to another (`b = a`), you are not creating a new object. You are simply attaching a **second label** to the same object in memory. This is called **aliasing**.

## Identity vs. Value
Every object in Python has three properties:
1.  **Type:** What the object can do (e.g., `list`, `int`).
2.  **Value:** The data stored in the object (e.g., `[1, 2, 3]`).
3.  **Identity:** A unique integer representing the object's address in memory (accessible via `id()`).

| Operator | Check | 
| :--- | :--- |
| `==` | Do these objects have the same **value**? |
| `is` | Are these names bound to the **same identity**? |

````python
# Two different objects with the same value
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # True (Same values)
print(a is b)  # False (Different identities in memory)

# Two names for the same object
c = a
print(c is a)  # True (They share the same identity)
````

## Careful: References are a frequent source of bugs
Because lists are **mutable**, modifying the object through one name will reflect across all its aliases.

```python
a = [1, 2, 3]
b = a

b.append(4) 
print(a)  # Output: [1, 2, 3, 4]
```

## Shallow vs. Deep Copying
If you need an independent version of a list, you must explicitly **copy** it. However, because lists can be nested (a list inside a list), we have two levels of copying:

### 1. Shallow Copy (`copy`)
Creates a new list object, but fills it with **references** to the items in the original list.
* **If the items are immutable (numbers, strings):** The copy behaves independently.
* **If the items are mutable (nested lists):** Both the original and the copy will still point to the same nested object.

### 2. Deep Copy (`deepcopy`)
Recursively creates a new object for the list **and** new objects for every item found inside it. The two datasets become entirely independent.


```python
from copy import copy, deepcopy

original = [1, 2, [3, 4]]

# A shallow copy still shares the inner [3, 4] list
shallow = copy(original)

# A deep copy creates a brand new inner [3, 4] list
deep = deepcopy(original)

# Modifying the nested list...
original[2].append(5)

print(f"Original: {original}") # [1, 2, [3, 4, 5]]
print(f"Shallow:  {shallow}")  # [1, 2, [3, 4, 5]] (Affected!)
print(f"Deep:     {deep}")     # [1, 2, [3, 4]]    (Safe!)
```

## Values and References

````python
list_a = [1, 1, 1]
list_b = list_a
list_c = [1, 1, 1]

print("=== " + "Initial setup" + " ===")

print(f"list_a == list_b: {list_a == list_b}")
print(f"list_a == list_c: {list_a == list_c}")
print(f"list_b == list_c: {list_b == list_c}")

list_b[0] = 99

print("=== " + "After assignment" + " ===")

print(f"list_a == list_b: {list_a == list_b}")
print(f"list_a == list_c: {list_a == list_c}")
print(f"list_b == list_c: {list_b == list_c}")

print("=== " + "Values" + " ===")

print(f"list_a: {list_a}")
print(f"list_b: {list_b}")
print(f"list_c: {list_c}")

print("=== " + "Identities" + " ===")

print(f"list_a is list_b: {list_a is list_b}")
print(f"list_b islist_c: {list_b is list_c}")
print(f"list_a is list_c: {list_a is list_c}")
````
